```python
import pandas as pd
import numpy as np
```

# Load the ratings by user information that you collected into a pandas dataframe.

```python
originalData = pd.read_csv('Week 7 - Aggregating Data.csv')
data = originalData.copy()
data
```

NaN values changed to 0 for readability.

```python
data.set_index('Name', inplace = True)
data.fillna('0')
```

# Show the average ratings for each user and each movie.

```python
usermean = data.mean(axis=1, skipna=True, numeric_only=True)
data['User Mean Ratings'] = usermean
moviemean = data.mean(axis=0, skipna=True, numeric_only=True)
data.loc['Mean Movie Rating'] = moviemean
movieOut = data.loc[['Mean Movie Rating']]
movieOut.drop('User Mean Ratings', inplace=True, axis=1)
data.fillna('0')
```

# Create a new pandas dataframe, with normalized ratings for each user. Again, show the average ratings for each user and each movie.

```python
normalData = (data - data.min()) / (data.max() - data.min())
normalData.fillna('0')
```

# Provide a text-based conclusion: explain what might be advantages and disadvantages of using normalized ratings instead of the actual ratings.

Normalizing the data will remove any outliers and put the data on a common scale.  By putting the data on a common sclae the original values are changed.

With my data 1 being the highest rating and 5 being the lowest, normalizing the data only confuses the viewer.

# [Extra credit] Create another new pandas dataframe, with standardized ratings for each user. Once again, show the average ratings for each user and each movie.

```python
standardData = ((data - data.mean()) / data.std())
standardData.fillna('0')
```
