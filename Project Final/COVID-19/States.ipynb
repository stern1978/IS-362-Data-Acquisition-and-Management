## Loads latest data

```python
import pandas as pd
import datetime as dt

d = dt.datetime.today()
today = d.strftime('%m-%d-%Y')
yesterday = d - dt.timedelta(days=1)
previous = yesterday.strftime('%m-%d-%Y')

url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/"

try:
    originalData = pd.read_csv(url + today + '.csv')

except:
    today = previous
    originalData = pd.read_csv(url + today + '.csv')
data = originalData.copy()
today
```

## NY state counties sorted by the number of Confirmed cases.

```python
ny = data.sort_values(by = 'Province_State').rename(columns={'Admin2':'County', 'Province_State':'State'})
ny.set_index('County', inplace=True)
ny[(ny['State'] == "New York")][['Confirmed', 'Deaths']].sort_values(by='Confirmed', ascending=False).head(10)
```

## NY state counties sorted by the number of Deaths.

```python
ny[(ny['State'] == "New York")][['Confirmed', 'Deaths']].sort_values(by='Deaths', ascending=False).head(10)
```

## States sorted by number of Confirmed cases.

```python
usa = data.sort_values(by = 'Province_State')
drate = usa['Deaths']/usa['Confirmed']
usa[usa['Country_Region'] == 'US'].groupby(['Province_State'])[['Confirmed', 'Deaths']].sum().sort_values(by=['Confirmed'], ascending = False).drop('Recovered')
```

## Top 10 Countries sorted by number of Confirmed cases of COVID-19

```python
country = data.sort_values(by='Country_Region')
country[['Country_Region', 
         'Confirmed', 
         'Recovered', 
         'Deaths', 
         'Active']].groupby(['Country_Region']).sum().sort_values(by = ['Country_Region']).sort_values(by = ['Active'], ascending = False).head(10)
```

```python
data['Total'] = data['Confirmed']
data[['Country_Region', 
      'Total', 
      'Recovered', 
      'Deaths']].groupby(['Country_Region']).sum().sort_values(by = ['Country_Region']).sort_values(by = ['Total', 'Recovered', 'Deaths'], ascending = True).head(10)
```

```python
data[['Country_Region', 
      'Confirmed', 
      'Recovered', 
      'Deaths']].groupby(['Country_Region']).sum().sort_values(by = ['Country_Region']).sort_values(by = ['Recovered', 'Deaths', 'Confirmed'], ascending = False).head(10)
```
